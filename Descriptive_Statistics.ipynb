```
# Use the describe() method on the range saleprice. What is the minimum value in the range?
prices_df['saleprice'].describe()

# Answer: 34,900
```

```
# Use the describe() method on the range lotarea. What is the mean value in the range?
prices_df['lotarea'].describe()

# Answer: 10,516
```

```
# Is there a statistically significant difference between the prices of houses that 
# have an open porch versus houses that do not have an open porch? How did you handle 
# null values?

# Checking the data
prices_df[['openporchsf', 'saleprice']].info()
# There are no nulls, assume '0' equates to no porch, calling homes that have a
# porch
with_open_porches = prices_df[['openporchsf', 'saleprice']].iloc[(prices_df['openporchsf'] > 0).values]
with_open_porches.info()
# Calling homes that don't have a porch
without_open_porches = prices_df[['openporchsf', 'saleprice']].iloc[(prices_df['openporchsf'] == 0).values]
without_open_porches.info()
# T-test and P-value results
stats.ttest_ind(with_open_porches['saleprice'], without_open_porches['saleprice'])

# Answer: Because the p-value is very small, you can conclude that this difference 
# is significant.
```

```
# Is there a significant correlation between the lotarea and the saleprice variables?
stats.pearsonr(prices_df['lotarea'], prices_df['saleprice'])

# Answer: They are positively correlated, and the relationship is statistically 
# significant.
```
